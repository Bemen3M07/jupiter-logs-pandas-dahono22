```python
Exercici 7: Informe final
Es crearà un document PDF amb ReportLab que inclourà:
Resultats de càlculs de l’Exercici 5.
Gràfics generats a l’Exercici 6.
Descripció de com es va implementar el logging.
```

```python
import pandas as pd
from matplotlib import pyplot as plt
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.platypus import Image

# **EXERCICI 5**
# Llegir dades
df = pd.read_csv('dades.csv')

# Mitjana per alumne
df['Mitjana'] = df.iloc[:, 1:].mean(axis=1)

# Percentatge d'aprovats
aprovats = (df['Mitjana'] >= 5).mean() * 100

# Notes més alta i baixa
nota_max = df.iloc[:, 1:].max().max()
nota_min = df.iloc[:, 1:].min().min()

resultats_text = (
    f"Mitjana global: {df['Mitjana'].mean():.2f}\n"
    f"Percentatge d'aprovats: {aprovats:.2f}%\n"
    f"Nota més alta: {nota_max}, Nota més baixa: {nota_min}\n"
)

# **EXERCICI 6**
# Guardar els gràfics com a imatges
data = pd.read_csv('dades.csv')

# Gràfica de barres
data.set_index('Nom')[['Nota_M01', 'Nota_M04', 'Nota_M03', 'Nota_M05']].plot(kind='bar', figsize=(10, 6))
plt.title('Notes per Mòdul')
plt.tight_layout()
plt.savefig('grafic_barres.png')
plt.close()

# Gràfica de línies
data.set_index('Nom').T.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Comparació de Notes entre Estudiants')
plt.tight_layout()
plt.savefig('grafic_linies.png')
plt.close()

# Gràfic circular
average_notes = data[['Nota_M01', 'Nota_M04', 'Nota_M03', 'Nota_M05']].mean()
average_notes.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', startangle=90)
plt.title('Distribució de les Notes Mitjanes per Mòdul')
plt.tight_layout()
plt.savefig('grafic_pastis.png')
plt.close()

# **GENERAR EL PDF**
c = canvas.Canvas("informe_final.pdf", pagesize=letter)

# Títol
c.drawString(100, 750, "Informe Final")

# Resultats de l'exercici 5
c.drawString(100, 700, "Resultats de l'Exercici 5:")
text = c.beginText(100, 680)
text.setFont("Helvetica", 10)
text.textLines(resultats_text)
c.drawText(text)

# Descripció del logging
logging_desc = (
    "El logging s'ha implementat per rastrejar i depurar el flux del programa.\n"
    "S'han registrat missatges en diferents nivells (INFO, ERROR) per a seguiment.\n"
)
c.drawString(100, 650, "Descripció del Logging:")
text = c.beginText(100, 630)
text.setFont("Helvetica", 10)
text.textLines(logging_desc)
c.drawText(text)

# Afegir els gràfics
c.drawString(100, 600, "Gràfics de l'Exercici 6:")
c.drawImage("grafic_barres.png", 100, 400, width=300, height=200)
c.drawImage("grafic_linies.png", 100, 180, width=300, height=200)

c.save()
```
