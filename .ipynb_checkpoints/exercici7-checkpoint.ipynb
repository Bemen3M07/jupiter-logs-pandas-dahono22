```python
Exercici 7: Informe final
Es crearà un document PDF amb ReportLab que inclourà:
Resultats de càlculs de l’Exercici 5.
Gràfics generats a l’Exercici 6.
Descripció de com es va implementar el logging.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
from reportlab.platypus import Image

# Paso 1: Resultados del ejercicio 5
# Leer datos
df = pd.read_csv('/mnt/data/dades.csv')

# Calcular métricas
df['Mitjana'] = df.iloc[:, 1:].mean(axis=1)
aprovats = (df['Mitjana'] >= 5).mean() * 100
nota_max = df.iloc[:, 1:].max().max()
nota_min = df.iloc[:, 1:].min().min()
mitjana_global = df['Mitjana'].mean()

# Paso 2: Gráficas del ejercicio 6
# Gráfica 1: Notas por módulo (barras)
plt.figure(figsize=(10, 6))
df.set_index('Nom')[['Nota_M01', 'Nota_M04', 'Nota_M03', 'Nota_M05']].plot(kind='bar')
plt.title('Notes per Mòdul')
plt.ylabel('Nota')
plt.xlabel('Estudiant')
plt.legend(title='Mòduls')
plt.tight_layout()
grafica1_path = '/mnt/data/grafica1.png'
plt.savefig(grafica1_path)
plt.close()

# Gráfica 2: Comparación de notas (líneas)
df.set_index('Nom').T.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Comparació de Notes entre Estudiants')
plt.ylabel('Nota')
plt.xlabel('Mòduls')
plt.grid(True)
plt.tight_layout()
grafica2_path = '/mnt/data/grafica2.png'
plt.savefig(grafica2_path)
plt.close()

# Gráfica 3: Relación entre Nota_M01 y Nota_M04 (dispersión)
plt.figure(figsize=(8, 6))
plt.scatter(df['Nota_M01'], df['Nota_M04'], color='purple', alpha=0.7, s=100)
plt.title('Relació entre Nota_M01 i Nota_M04')
plt.xlabel('Nota_M01')
plt.ylabel('Nota_M04')
plt.grid(True)
plt.tight_layout()
grafica3_path = '/mnt/data/grafica3.png'
plt.savefig(grafica3_path)
plt.close()

# Gráfica 4: Promedios por módulo (circular)
average_notes = df[['Nota_M01', 'Nota_M04', 'Nota_M03', 'Nota_M05']].mean()
average_notes.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(8, 8),
                   labels=average_notes.index, colors=['gold', 'skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribució de les Notes Mitjanes per Mòdul')
plt.ylabel('')  # Eliminar etiqueta del eje Y
plt.tight_layout()
grafica4_path = '/mnt/data/grafica4.png'
plt.savefig(grafica4_path)
plt.close()

# Paso 3: Generar el PDF
pdf_path = "/mnt/data/informe_final.pdf"
c = canvas.Canvas(pdf_path, pagesize=letter)

# Título
c.setFont("Helvetica-Bold", 16)
c.drawString(100, 750, "Informe Final")

# Apartado 1: Resultados del Ejercicio 5
c.setFont("Helvetica-Bold", 14)
c.drawString(100, 720, "1. Resultats de càlculs de l’Exercici 5.")
c.setFont("Helvetica", 12)
c.drawString(100, 700, f"- Mitjana global: {mitjana_global:.2f}")
c.drawString(100, 680, f"- Percentatge d'aprovats: {aprovats:.2f}%")
c.drawString(100, 660, f"- Nota més alta: {nota_max}, Nota més baixa: {nota_min}")

# Apartado 2: Gráficas del Ejercicio 6
c.setFont("Helvetica-Bold", 14)
c.drawString(100, 640, "2. Gràfics generats a l’Exercici 6.")
c.drawImage(grafica1_path, 100, 400, width=400, height=200)
c.drawImage(grafica2_path, 100, 200, width=400, height=200)
c.showPage()
c.drawImage(grafica3_path, 100, 500, width=400, height=200)
c.drawImage(grafica4_path, 100, 300, width=400, height=200)

# Apartado 3: Descripción del Logging
c.setFont("Helvetica-Bold", 14)
c.drawString(100, 260, "3. Descripció de com es va implementar el logging.")
logging_description = """
Es va utilitzar la llibreria logging per registrar informació del programa.
Això permet controlar missatges d’informació, advertiments i errors.
Exemple de codi utilitzat:
"""
c.setFont("Helvetica", 12)
c.drawString(100, 240, logging_description)
logging_code = """
import logging

logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')
logging.info('Missatge informatiu.')
logging.warning('Advertiment.')
logging.error('Error.')
"""
c.drawString(100, 220, logging_code.strip())

# Guardar PDF
c.save()
pdf_path
```
