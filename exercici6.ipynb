```python
Exercici 6: Gràfics
Faré servir Matplotlib per generar gràfics visuals:

Histogrames de notes.
Gràfics de barres per mitjanes.
Pastís per percentatges.
Línies per mostrar l'evolució de notes.
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# Llegir el fitxer dades.csv
data = pd.read_csv('dades.csv')

# Mostra les primeres files del conjunt de dades
print(data.head())

# Gràfica de barres per veure les notes de cada estudiant
data.set_index('Nom')[['Nota_M01', 'Nota_M04', 'Nota_M03', 'Nota_M05']].plot(kind='bar', figsize=(10, 6))
plt.title('Notes per Mòdul')
plt.ylabel('Nota')
plt.xlabel('Estudiant')
plt.xticks(rotation=0)
plt.legend(title='Mòduls')
plt.tight_layout()
plt.show()

# Gràfica de línies per comparar les notes dels estudiants
data.set_index('Nom').T.plot(kind='line', figsize=(10, 6), marker='o')
plt.title('Comparació de Notes entre Estudiants')
plt.ylabel('Nota')
plt.xlabel('Mòduls')
plt.legend(title='Estudiants')
plt.grid(True)
plt.tight_layout()
plt.show()

# Gràfic de dispersió: Relació entre Nota_M01 i Nota_M04
plt.figure(figsize=(8, 6))
plt.scatter(data['Nota_M01'], data['Nota_M04'], color='purple', alpha=0.7, s=100)
plt.title('Relació entre Nota_M01 i Nota_M04')
plt.xlabel('Nota_M01')
plt.ylabel('Nota_M04')
plt.grid(True)
plt.tight_layout()
plt.show()

# Gràfic circular amb la mitjana de les notes per mòdul
average_notes = data[['Nota_M01', 'Nota_M04', 'Nota_M03', 'Nota_M05']].mean()
average_notes.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', startangle=90, labels=average_notes.index, colors=['gold', 'skyblue', 'lightcoral', 'lightgreen'])
plt.title('Distribució de les Notes Mitjanes per Mòdul')
plt.ylabel('')  # Eliminar etiqueta del eix Y
plt.tight_layout()
plt.show()
```
